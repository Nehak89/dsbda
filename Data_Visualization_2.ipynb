```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.simplefilter('ignore')
```

```python
#Were younger passengers more likely to survive?
ax=df.boxplot(column='Age',by='Survived',figsize=(10,5))
ax.set_ylabel('Age')
ax.set_title('Boxplot:Survived VS Age')
```

```python
#Were younger passengers more likely to survive considering the sex?
ax=df.boxplot(column='Age',by=['Survived','Sex'],figsize=(10,5))
ax.set_ylabel('Age')
ax.set_title('Boxplot:Survived VS Age')
```

```python
#Did survival rate differ by boarding the location o& age?  
ax=df.boxplot(column='Age',by=['Survived','Embarked'],figsize=(10,5))
ax.set_ylabel('Age')
ax.set_title('Boxplot:Embarked VS Age')
```

```python
#Were high class passengers more likely to survive?
ax=df.boxplot(column='Pclass',by='Survived',figsize=(10,5))
ax.set_ylabel('Pclass')
ax.set_title('Boxplot:Pclass VS Age')
```

```python
#Did passngers who paid high survived?
ax=df.boxplot(column='Fare',by='Survived',figsize=(10,5))
ax.set_ylabel('Fare')
ax.set_title('Boxplot:Fare VS Age')
```
