```python
for c in nc:
    Q1=df[c].quantile(0.25)
    Q3=df[c].quantile(0.75)
    IQR=Q3-Q1
    lowerbound=Q1-1.5*IQR
    upperbound=Q3+1.5*IQR
    df[c]=np.where((df[c]<lowerbound)|(df[c]>upperbound),df[c].median(),df[c])
df
```
