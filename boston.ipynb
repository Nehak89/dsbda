```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
```

```python
df['CRIM'].fillna(df['CRIM'].mean(), inplace=True)
df['ZN'].fillna(df['ZN'].mean(), inplace=True)
df['INDUS'].fillna(df['INDUS'].mean(), inplace=True)
df['CHAS'].fillna(df['CHAS'].mean(), inplace=True)
df['AGE'].fillna(df['AGE'].mean(), inplace=True)
df['LSTAT'].fillna(df['LSTAT'].mean(), inplace=True)
```

```python
df.isnull().value_counts()
```

```python
df.isnull().sum()
```

```python
x=df.drop('MEDV',axis=1)
y=df['MEDV']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8,random_state=10)
```

```python
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
print("Mean Squared Error:",mean_squared_error(y_test,y_pred)) #average squared difference between the predicted values and the actual values
```

```python
print("r2_score:",r2_score(y_test,y_pred)) # how well your model explains the variance in the target variable
```
